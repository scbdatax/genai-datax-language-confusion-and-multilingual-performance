```python
import pandas as pd
df = pd.read_parquet("hf://datasets/airesearch/WangchanThaiInstruct_7.24/data/train-00000-of-00001.parquet")
```

```python
df_batch_1 = pd.read_excel("../input/wti_clean.xlsx")
df_batch_1 = df_batch_1[df_batch_1['Task_type'] == 'Summarization']
df = df.merge(df_batch_1["ID"], on="ID", how="inner")
```

## Task: Summarization

```python
sum_df = df[df['Task_type'] == 'Summarization']
```

```python
sum_df.head()
```

```python
len(sum_df)
```

```python
# Drop columns
sum_df = sum_df.drop(columns=['Tags', 'Task_type', 'License'])
```

```python
sum_df = sum_df.rename(
    columns={
        'Instruction': 'question',
        'Input': 'context',
        'Output': 'answer',
		'Domain': 'subject'
	}
)
```

```python
sum_df.to_json("../output/WTI_SUM/WTI_SUM_questions_gt_th.jsonl")
```
