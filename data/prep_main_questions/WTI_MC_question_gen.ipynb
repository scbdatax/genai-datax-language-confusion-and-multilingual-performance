```python
answers = []
for i in range(len(multiple_choices_df)):
	prompt = multiple_choices_df.iloc[i]['Output'] + '\n จากเฉลยข้างต้น คำตอบที่ถูกต้องคือข้อไหน ตอบแค่ตัวอักษร เช่น A หากมีหลายคำตอบ ให้แยกด้วย , เช่น A, B, C'
	res = gpt.get_response(prompt)
	answers.append(res)
```
