```python
import pandas as pd
df = pd.read_parquet("hf://datasets/airesearch/WangchanThaiInstruct_7.24/data/train-00000-of-00001.parquet")
```

## Task: Closed QA

```python
cqa_df = df[df['Task_type'] == 'Closed QA']
```

```python
cqa_df.head()
```

```python
len(cqa_df)
```

```python
# Drop columns
cqa_df = cqa_df.drop(columns=['Thai_Specfic', 'Tags', 'Task_type', 'License'])
```

```python
cqa_df = cqa_df.rename(
    columns={
        'Instruction': 'question',
        'Input': 'context',
        'Output': 'answer',
		'Domain': 'subject'
	}
)
```

```python
cqa_df.to_json("../output/WTI_CQA/WTI_CQA_questions_gt_th.jsonl")
```

### English version
