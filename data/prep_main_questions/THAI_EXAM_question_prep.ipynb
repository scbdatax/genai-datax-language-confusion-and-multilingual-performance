```python
from pandas import concat
from pandas import read_json
from pandas import DataFrame


base_path = "hf://datasets/scb10x/thai_exam/"
exams = [
    "a_level",
    "ic",
    "onet",
    "tgat",
    "tpat1",
]
df = DataFrame()

for exam in exams:
    splits = {
        "train": f"data/{exam}/{exam}_train.jsonl", 
        "test": f"data/{exam}/{exam}_test.jsonl",
    }
    tmp_train = read_json(base_path + splits["train"], lines=True)
    tmp_test = read_json(base_path + splits["test"], lines=True)

    tmp_train["split"] = "train"
    tmp_test["split"] = "test"
    tmp_data = concat(
        objs=[tmp_train, tmp_test],
        ignore_index=True,
    )
    tmp_data["ID"] = [
        f"{exam}_{i:03d}" for i in range(1, len(tmp_data)+1)
    ]
    tmp_data["exam"] = exam

    df = concat(
        objs=[df, tmp_data],
        ignore_index=True,
    )
```
